```python
import pandas as pd
```

```python
df = pd.read_csv("../data/data_unclean.csv", sep=";")
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

```python
df = df.drop_duplicates(subset=["recipe_name", "type_of_dish", "main_ingredient", "cuisine"])
df.shape
```

```python
df.insert(0, "id", df.index)
df.head()
```

```python
df.to_csv("../data/recipe.csv", index=False)
```
